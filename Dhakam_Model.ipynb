```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

seed = 1
numInstance = 200
np.random.seed(seed)  # Use np.random.seed(seed) to set the seed

X = np.random.rand(numInstance, 1).reshape(-1, 1)
X2 = 0.5 * X + np.random.normal(0, 0.04, size=numInstance).reshape(-1, 1)
X3 = 0.5 * X2 + np.random.normal(0, 0.04, size=numInstance).reshape(-1, 1)
X4 = 0.5 * X3 + np.random.normal(0, 0.04, size=numInstance).reshape(-1, 1)
X5 = 0.5 * X4 + np.random.normal(0, 0.04, size=numInstance).reshape(-1, 1)

Y_true = -3 * X + 1
Y = Y_true + np.random.normal(size=numInstance).reshape(-1, 1)

numTrain = 20
numTest = numInstance - numTrain

X_Train = X[:-numTest]
X_Test = X[-numTest:]
Y_Train = Y[:-numTest]
Y_Test = Y[-numTest:]

X_Train2 = np.column_stack([X[:-numTest], X2[:-numTest]])
X_Test2 = np.column_stack([X[-numTest:], X2[-numTest:]])
X_Train3 = np.column_stack([X2[:-numTest], X3[:-numTest]])
X_Test3 = np.column_stack([X2[-numTest:], X3[-numTest:]])
X_Train4 = np.column_stack([X3[:-numTest], X4[:-numTest]])
X_Test4 = np.column_stack([X3[-numTest:], X4[-numTest:]])
X_Train5 = np.column_stack([X4[:-numTest], X5[:-numTest]])
X_Test5 = np.column_stack([X4[-numTest:], X5[-numTest:]])

# Create and train the linear regression models
regr = linear_model.LinearRegression()
regr.fit(X_Train, Y_Train)

regr2 = linear_model.LinearRegression()
regr2.fit(X_Train2, Y_Train)

regr3 = linear_model.LinearRegression()
regr3.fit(X_Train3, Y_Train)

regr4 = linear_model.LinearRegression()
regr4.fit(X_Train4, Y_Train)

regr5 = linear_model.LinearRegression()
regr5.fit(X_Train5, Y_Train)

# Make predictions
Y_pred_Train = regr.predict(X_Train)
Y_pred_Test = regr.predict(X_Test)

Y_pred_Train2 = regr2.predict(X_Train2)
Y_pred_Test2 = regr2.predict(X_Test2)

Y_pred_Train3 = regr3.predict(X_Train3)
Y_pred_Test3 = regr3.predict(X_Test3)

Y_pred_Train4 = regr4.predict(X_Train4)
Y_pred_Test4 = regr4.predict(X_Test4)

Y_pred_Train5 = regr5.predict(X_Train5)
Y_pred_Test5 = regr5.predict(X_Test5)

# Create a list of models and their metrics
models = [
    ('Model 1', Y_pred_Train, Y_pred_Test, regr.coef_[0], regr.intercept_[0]),
    ('Model 2', Y_pred_Train2, Y_pred_Test2, regr2.coef_[0], regr2.intercept_[0]),
    ('Model 3', Y_pred_Train3, Y_pred_Test3, regr3.coef_[0], regr3.intercept_[0]),
    ('Model 4', Y_pred_Train4, Y_pred_Test4, regr4.coef_[0], regr4.intercept_[0]),
    ('Model 5', Y_pred_Train5, Y_pred_Test5, regr5.coef_[0], regr5.intercept_[0])
]

# Print the metrics for each model
for model in models:
    model_name, train_error, test_error, coef, intercept = model
    train_rmse = np.sqrt(mean_squared_error(Y_Train, train_error))
    test_rmse = np.sqrt(mean_squared_error(Y_Test, test_error))
    sum_of_attributes = np.abs(coef).sum() + np.abs(intercept)
    
    print(f"Model: {model_name}")
    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test RMSE: {test_rmse:.4f}")
    print(f"Sum of Absolute Attributes: {sum_of_attributes:.4f}\n")
```
